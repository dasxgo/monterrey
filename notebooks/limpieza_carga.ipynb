```python
import pandas as pd
import os
```

```python
df_gastos_varios = pd.read_excel('/content/12_DICIEMBRE_2020.xlsx', sheet_name='GASTOS VARIOS', skiprows=5)
df_gastos_varios['SHEET']='GASTOS VARIOS'
```

```python
df_contratistas = pd.read_excel('/content/12_DICIEMBRE_2020.xlsx', sheet_name='CONTRATISTAS Y FDO FED', skiprows=5)
df_contratistas['SHEET']='CONTRATISTAS Y FDO FED'
```

```python
df_servicios = pd.read_excel('/content/12_DICIEMBRE_2020.xlsx', sheet_name='SERV PPROF', skiprows=5)
df_servicios['SHEET']='SERV PPROF'
```

```python
df_comunicacion = pd.read_excel('/content/12_DICIEMBRE_2020.xlsx', sheet_name='COMUNICACION', skiprows=5)
df_comunicacion['SHEET']='COMUNICACION'
```
